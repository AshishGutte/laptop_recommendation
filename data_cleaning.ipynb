```python
# =========================
# Price Cleaning
# =========================
df['Price'] = (
    df['Price'].astype(str)
    .str.extract(r'₹([\d,]+)')[0]
    .str.replace(',', '', regex=True)
    .astype(float)
)
df = df.dropna(subset=['Price'])
df['Price'] = df['Price'].astype(int)
```

```python
# =========================
# Remove 'Not found'  prices
# =========================
df = df[~(df['Price'] == 'Not found')]
```

```python
df
```

```python

def extract_laptop_name(text):
    return re.split(r'Intel|AMD|\(|-', text)[0].strip()

df['Clean Name'] = df['Product Name'].apply(extract_laptop_name)
df['Product Name'] = df['Clean Name']
df.drop(columns=['Clean Name'], inplace=True)
```

```python
df
```

```python

df['Brand'] = df['Product Name'].apply(lambda x: x.split()[0])
```

```python

df["Specifications"] = df["Specifications"].astype(str).apply(
    lambda x: re.sub(r"(?<=[a-zA-Z])(?=\d)", " ", x)
)
```

```python

def extract_feature(pattern, text):
    match = re.search(pattern, text)
    return match.group(0).strip() if match else "Not found"

df["Processor"] = df["Specifications"].apply(lambda x: extract_feature(r"(Intel|AMD|Celeron)[^,\n]*Processor\s*\([^)]*\)?", x))
df["RAM"] = df["Specifications"].apply(lambda x: extract_feature(r"\d+\s*GB", x))
df["DDR"] = df["Specifications"].apply(lambda x: extract_feature(r"(LP)?DDR\d?", x))
df["OS"] = df["Specifications"].apply(lambda x: extract_feature(r"Windows\s+11\s*(Home)?", x))
df["Storage"] = df["Specifications"].apply(lambda x: extract_feature(r"\d+\s*(GB|TB)\s*SSD", x))
df["Display"] = df["Specifications"].apply(lambda x: extract_feature(r"\d{2}\.\d{2}\s*cm\s*\([^)]+\)", x))
df["Warranty"] = df["Specifications"].apply(lambda x: extract_feature(r"\d+\s*Year\s+(Onsite|Carry-in)?\s*Warranty", x))
```

```python


def format_product(row):
    return (
        f"Product: {row['Product Name']}, Brand: {row['Brand']}, Rating: {row['Rating']}, Price: ₹{row['Price']}, "
        f"Processor: {row['Processor']}, RAM: {row['RAM']} GB, DDR: {row['DDR']}, OS: {row['OS']}, "
        f"Storage: {row['Storage']}, "
        f"Weight: {row['Weight']} kg, "
        f"Warranty: {row['Warranty']}"
    )
df["text"] = df.apply(format_product, axis=1)
```

```python
import faiss
import numpy as np
import json
from sentence_transformers import SentenceTransformer
from langchain_community.chat_models import ChatOllama
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Load FAISS index and metadata
index = faiss.read_index("laptop_index.faiss")
with open("laptop_metadata.json") as f:
    metadata = json.load(f)

# Load embedding model
# model = SentenceTransformer("all-MiniLM-L6-v2")
model = SentenceTransformer("all-mpnet-base-v2")


# Retrieve top K similar products
def retrieve_products(query, top_k=3):
    query_embedding = model.encode([query])
    faiss.normalize_L2(query_embedding)
    D, I = index.search(np.array(query_embedding, dtype=np.float32), top_k)
    return [metadata[i] for i in I[0]]

# Load LLM
llm = ChatOllama(model="llama3.2")

# Prompt Template
template = """
You are a helpful product recommendation assistant.

ONLY use the product information provided below to answer the user's query.
Do NOT make up or hallucinate any specs or features.

User Query:
"{query}"

Product Information:
{context}

Instructions:
1. Understand the user's needs (e.g., budget, use case like gaming, coding, work-from-home).
2. STRICTLY filter out products that exceed the budget mentioned in the user query.
3. Among the remaining options, find and recommend the best product(s) based on key specs:
   - RAM
   - Storage
   - Processor
   - Graphics (if mentioned in the query)
4. If no products meet the use case AND fall within the budget, politely inform the user and suggest trying a higher budget.
5. In your recommendation, clearly mention:
   - Product Name
   - Price
   - RAM, Storage, Processor 
   - Flipkart Product URL
   - Short explanation of why it's a good match or why no match was found.

Respond in a user-friendly, clear, and concise manner.
"""

prompt = PromptTemplate(input_variables=["query", "context"], template=template)
chain = LLMChain(llm=llm, prompt=prompt)

# Generate final answer
def generate_answer(query, top_k=5):
    results = retrieve_products(query, top_k=top_k)

    # Format full context with key features (safe from hallucination)
    context = "\n".join(
        f"- Name: {r.get('Product Name', '')}\n"
        f"  Price: ₹{r.get('Price', '')}\n"
        f"  RAM: {r.get('RAM', '')}\n"
        f"  Storage: {r.get('Storage', '')}\n"
        f"  Processor: {r.get('Processor', '')}\n"
        f"  Graphics: {r.get('Specifications', '')}\n"
        f"  URL: {r.get('product_url', '')}"
        for r in results
    )

    answer = chain.run(query=query, context=context)
    return answer, results
```

```python
# User query
query = "i want a mobile for office work under 100000"

# Generate answer and retrieve product results
answer, products = generate_answer(query)

# Print LLM answer
print("📢 Answer:\n")
print(answer)

# Print top retrieved products with key details
print("\n🛒 Retrieved Products:\n")
for p in products:
    print(
        f"- Name: {p.get('product_name', '')}\n"
        f"  Price: ₹{p.get('price', '')}\n"
        f"  RAM: {p.get('ram', '')}\n"
        f"  Storage: {p.get('storage', '')}\n"
        f"  Processor: {p.get('processor', '')}\n"
        f"  URL: {p.get('url', '')}\n"
    )
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import uuid
import json

# Load CSV data
df = pd.read_csv("flipkart_laptop_cleaned.csv")

# Load embedding model
model = SentenceTransformer("all-mpnet-base-v2")

# Generate embeddings from the 'text' column
texts = df["text"].astype(str).tolist()
embeddings = model.encode(texts, show_progress_bar=True)

# Set up FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

# Normalize and add embeddings
faiss.normalize_L2(embeddings)
index.add(np.array(embeddings, dtype=np.float32))

# Save FAISS index
faiss.write_index(index, "laptop_index.faiss")

# Prepare metadata (all lowercase, underscore format for consistency)
metadata = []
for _, row in df.iterrows():
    metadata.append({
        "id": str(uuid.uuid4()),
        "product_name": row.get("Product Name", ""),
        "price": row.get("Price", ""),
        "rating": row.get("Rating", ""),
        "processor": row.get("Processor", ""),
        "ram": row.get("RAM", ""),
        "os": row.get("OS", ""),
        "storage": row.get("Storage", ""),
        "display": row.get("Display", ""),
        "warranty": row.get("Warranty", ""),
        "ddr": row.get("DDR", ""),
        "specs": row.get("Specifications", ""),
        "url": row.get("Product URL", ""),
        "battery_life": row.get("Battery Life", ""),
        "weight": row.get("Weight", ""),
        "webcam": row.get("Webcam", ""),
        "display_size": row.get("Display Size", ""),
        "brand": row.get("Brand", ""),
        "text": row.get("text", "")
    })

# Save metadata
with open("laptop_metadata.json", "w") as f:
    json.dump(metadata, f)

print("✅ All products embedded and stored in FAISS with full metadata.")
```
